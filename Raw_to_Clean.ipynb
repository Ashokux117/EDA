- *we remove regular expression in categorical datatype like Name column*

```python
# here we treat regular expressions 
data['Name']= data['Name'].str.replace('[^a-zA-Z]','',regex=True)
```

- *here we check categorical missing values and fill missing values*

```python
data["Location"].isnull()
```

```python
data["Location"]
```

-  *first we remove all regular expression in  the numerical column and check misssing values and then we fill missing velues* 

```python
data["Age"]
```

```python
# here we remove reular exp
data["Age"]=data["Age"].str.extract('(\d+)')
```

```python

data["Age"]
```

```python
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
```

```python
# fill misssing values using mean
data['Age'].fillna(data['Age'].mean(), inplace=True)
```

```python
# convert to dtype float to int
data['Age'] = data['Age'].round().astype(int)
```

- *sucessfully cleaned age column*
